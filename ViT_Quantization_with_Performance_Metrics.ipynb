```
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
import torch.ao.quantization.quantize_fx as quantize_fx
from torch.ao.quantization import QConfigMapping
import timm
import os
import time
import copy
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ==============================================================================
# 1. Setup and Helper Functions
# ==============================================================================

def get_cifar100_dataloaders(train_batch_size=64, test_batch_size=100):
    """
    Prepares CIFAR-100 dataloaders with transformations suitable for ViT.
    """
    # ViT models were pre-trained on 224x224 images.
    # We need to resize and normalize the CIFAR-100 images accordingly.
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.5071, 0.4867, 0.4408], std=[0.2675, 0.2565, 0.2761]),
    ])

    trainset = torchvision.datasets.CIFAR100(root='./data', train=True,
                                             download=True, transform=transform)
    # Use a subset for faster demonstration
    train_subset_indices = list(range(0, len(trainset), 20)) # Use 5% of the data
    train_subset = torch.utils.data.Subset(trainset, train_subset_indices)

    trainloader = DataLoader(train_subset, batch_size=train_batch_size,
                             shuffle=True, num_workers=2)

    testset = torchvision.datasets.CIFAR100(root='./data', train=False,
                                            download=True, transform=transform)
    # Use a subset for faster evaluation
    test_subset_indices = list(range(0, len(testset), 20)) # Use 5% of the data
    test_subset = torch.utils.data.Subset(testset, test_subset_indices)

    testloader = DataLoader(test_subset, batch_size=test_batch_size,
                            shuffle=False, num_workers=2)

    # Dataloader for calibration (a few batches from training set)
    calibration_loader = DataLoader(train_subset, batch_size=16, shuffle=True, num_workers=2)

    return trainloader, testloader, calibration_loader

def evaluate_model(model, dataloader, device):
    """
    Evaluates the model's accuracy on the given dataloader.
    """
    model.to(device)
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for data in dataloader:
            images, labels = data
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    accuracy = 100 * correct / total
    print(f"Accuracy: {accuracy:.2f}%")
    return accuracy

def print_model_size(model, label):
    """
    Calculates and returns the size of the model's state dictionary in MB.
    """
    if not os.path.exists("temp_models"):
        os.makedirs("temp_models")
    temp_path = "temp_models/temp.p"
    torch.save(model.state_dict(), temp_path)
    size_mb = os.path.getsize(temp_path) / 1e6
    print(f"Size of {label} model: {size_mb:.2f} MB")
    os.remove(temp_path)
    return size_mb

def measure_inference_latency(model, device, dummy_input):
    """
    Measures the average inference latency of the model.
    """
    model.to(device)
    model.eval()
    dummy_input = dummy_input.to(device)

    # Warmup runs to stabilize timings
    for _ in range(10):
        _ = model(dummy_input)

    # Measurement runs
    timings = np.zeros((100,))
    with torch.no_grad():
        for i in range(100):
            start_time = time.time()
            _ = model(dummy_input)
            end_time = time.time()
            # --- MODIFICATION: Correctly append the timing for each run ---
            timings[i] = (end_time - start_time) * 1000 # Store in milliseconds

    # --- MODIFICATION: Calculate mean of all timings ---
    mean_latency_ms = np.mean(timings)
    print(f"Average inference latency: {mean_latency_ms:.3f} ms")
    return mean_latency_ms

# ==============================================================================
# 2. Model Definition and Workflows
# ==============================================================================

def create_vit_model(num_classes=100): # Updated for CIFAR-100
    """
    Loads a pre-trained ViT model and adapts its head for CIFAR-100.
    """
    # Using a smaller ViT model for faster execution
    model = timm.create_model('vit_tiny_patch16_224', pretrained=True)
    model.head = nn.Linear(model.head.in_features, num_classes)
    return model

def fp32_baseline_workflow(model, trainloader, testloader, device):
    """
    Fine-tunes and evaluates the FP32 model to establish a baseline.
    """
    print("--- Starting FP32 Baseline Workflow ---")
    model.to(device)

    criterion = nn.CrossEntropyLoss()
    # Using AdamW which often works better for transformers
    optimizer = optim.AdamW(model.parameters(), lr=1e-3, weight_decay=1e-2)

    # --- MODIFICATION: Added a learning rate scheduler ---
    # This will decrease the learning rate over time, helping training stability.
    scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.1)

    # --- MODIFICATION: Increased number of epochs to 30 ---
    num_epochs = 30
    print(f"Fine-tuning FP32 model for {num_epochs} epochs...")
    model.train()

    for epoch in range(num_epochs):
        for i, (inputs, labels) in enumerate(trainloader):
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

        # Call the scheduler after each epoch
        scheduler.step()

        # Print loss and current learning rate at the end of each epoch
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}, LR: {scheduler.get_last_lr()[0]:.6f}")

    print("Fine-tuning finished.")

    if not os.path.exists("models"):
        os.makedirs("models")
    torch.save(model.state_dict(), "models/vit_fp32_finetuned_cifar100.pth")
    print("FP32 fine-tuned model saved.")
    return model

def ptsq_workflow(fp32_model_path, calibration_loader):
    """
    Performs Post-Training Static Quantization using FX Graph Mode.
    """
    print("\n--- Starting PTSQ Workflow ---")

    model_fp32 = create_vit_model()
    model_fp32.load_state_dict(torch.load(fp32_model_path))
    model_fp32.to("cpu")
    model_fp32.eval()

    example_inputs = (torch.randn(1, 3, 224, 224),)
    qconfig_mapping = torch.ao.quantization.get_default_qconfig_mapping("x86")

    print("Preparing model for calibration...")
    model_prepared = quantize_fx.prepare_fx(model_fp32, qconfig_mapping, example_inputs)

    print("Calibrating model...")
    with torch.no_grad():
        for i, (calib_data, _) in enumerate(calibration_loader):
            if i >= 10:
                break
            model_prepared(calib_data)
    print("Calibration finished.")

    print("Converting model to INT8...")
    model_int8_static = quantize_fx.convert_fx(model_prepared)
    print("Conversion finished.")

    return model_int8_static

def qat_workflow(fp32_model_path, trainloader):
    """
    Performs Quantization-Aware Training using FX Graph Mode.
    """
    print("\n--- Starting QAT Workflow ---")

    model_to_qat = create_vit_model()
    model_to_qat.load_state_dict(torch.load(fp32_model_path))
    model_to_qat.to("cpu")

    example_inputs = (torch.randn(1, 3, 224, 224),)
    qconfig_mapping = torch.ao.quantization.get_default_qat_qconfig_mapping("x86")
    model_prepared = quantize_fx.prepare_qat_fx(model_to_qat, qconfig_mapping, example_inputs)

    print("Warming up QAT observers...")
    device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
    model_prepared.to(device)
    model_prepared.eval() # Set to eval mode for warm-up
    with torch.no_grad():
        for i, (inputs, _) in enumerate(trainloader):
            if i >= 10: # Use a few batches for warm-up
                break
            inputs = inputs.to(device)
            model_prepared(inputs)
    print("Observer warm-up finished.")

    print("Fine-tuning model with QAT...")
    model_prepared.train() # Set back to train mode

    criterion = nn.CrossEntropyLoss()
    optimizer = optim.SGD(model_prepared.parameters(), lr=1e-4)

    num_qat_epochs = 10
    for epoch in range(num_qat_epochs):
        for i, (inputs, labels) in enumerate(trainloader):
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model_prepared(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            if (i + 1) % 1 == 0:
                print(f"QAT Epoch [{epoch+1}/{num_qat_epochs}], Step [{i+1}/{len(trainloader)}], Loss: {loss.item():.4f}")
    print("QAT fine-tuning finished.")

    print("Converting QAT model to INT8...")
    model_prepared.to("cpu")
    model_int8_qat = quantize_fx.convert_fx(model_prepared)
    print("Conversion finished.")

    return model_int8_qat

# ==============================================================================
# 6. Main Execution and Visualization Block
# ==============================================================================

def plot_metrics(results_df):
    """
    Plots the comparison of Accuracy, Model Size, and Latency.
    """
    fig, axes = plt.subplots(1, 3, figsize=(20, 6))
    fig.suptitle('ViT Quantization Performance Comparison (CIFAR-100)', fontsize=16)
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

    # Accuracy Plot
    axes[0].bar(results_df.index, results_df['Accuracy (%)'], color=colors)
    axes[0].set_title('Accuracy Comparison')
    axes[0].set_ylabel('Accuracy (%)')
    axes[0].set_ylim(bottom=max(0, results_df['Accuracy (%)'].min() - 10))

    # Model Size Plot
    axes[1].bar(results_df.index, results_df['Model Size (MB)'], color=colors)
    axes[1].set_title('Model Size Comparison')
    axes[1].set_ylabel('Size (MB)')

    # Latency Plot
    axes[2].bar(results_df.index, results_df['Latency (ms)'], color=colors)
    axes[2].set_title('Inference Latency Comparison (CPU)')
    axes[2].set_ylabel('Latency (ms)')

    for ax in axes:
        ax.tick_params(axis='x', rotation=15)
        for container in ax.containers:
            ax.bar_label(container, fmt='%.2f', label_type='edge')

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()


if __name__ == '__main__':
    # Set device
    train_device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
    eval_device = torch.device("cpu") # Quantization is CPU-focused
    print(f"Using training device: {train_device}")
    print(f"Using evaluation device: {eval_device}")

    train_loader, test_loader, calib_loader = get_cifar100_dataloaders()

    # --- FP32 Baseline ---
    fp32_model_path = "models/vit_fp32_finetuned_cifar100.pth"
    fp32_model = create_vit_model()
    if not os.path.exists(fp32_model_path):
        fp32_model = fp32_baseline_workflow(fp32_model, train_loader, test_loader, train_device)
    else:
        print("--- Loading existing fine-tuned FP32 model ---")
        fp32_model.load_state_dict(torch.load(fp32_model_path))

    # --- Evaluate All Models ---
    results = {}
    dummy_input = torch.randn(1, 3, 224, 224)

    print("\n" + "="*50)
    print("--- Evaluating FP32 Model ---")
    fp32_accuracy = evaluate_model(fp32_model, test_loader, eval_device)
    fp32_size = print_model_size(fp32_model, "FP32")
    fp32_latency = measure_inference_latency(fp32_model, eval_device, dummy_input)
    results['FP32'] = {'Accuracy (%)': fp32_accuracy, 'Model Size (MB)': fp32_size, 'Latency (ms)': fp32_latency}

    # --- PTSQ Workflow ---
    ptsq_model = ptsq_workflow(fp32_model_path, calib_loader)
    print("\n" + "="*50)
    print("--- Evaluating PTSQ INT8 Model ---")
    ptsq_accuracy = evaluate_model(ptsq_model, test_loader, eval_device)
    ptsq_size = print_model_size(ptsq_model, "PTSQ INT8")
    ptsq_latency = measure_inference_latency(ptsq_model, eval_device, dummy_input)
    results['PTSQ INT8'] = {'Accuracy (%)': ptsq_accuracy, 'Model Size (MB)': ptsq_size, 'Latency (ms)': ptsq_latency}

    # --- QAT Workflow ---
    qat_model = qat_workflow(fp32_model_path, train_loader)
    print("\n" + "="*50)
    print("--- Evaluating QAT INT8 Model ---")
    qat_accuracy = evaluate_model(qat_model, test_loader, eval_device)
    qat_size = print_model_size(qat_model, "QAT INT8")
    qat_latency = measure_inference_latency(qat_model, eval_device, dummy_input)
    results['QAT INT8'] = {'Accuracy (%)': qat_accuracy, 'Model Size (MB)': qat_size, 'Latency (ms)': qat_latency}

    # --- Display Summary and Plots ---
    results_df = pd.DataFrame.from_dict(results, orient='index')
    print("\n" + "="*60)
    print(" " * 20 + "PERFORMANCE SUMMARY (CIFAR-100)")
    print("="*60)
    print(results_df)
    print("="*60)

    plot_metrics(results_df)
```
